## Задание №1

Задание 1
Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:

phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
Результат:
Фраза 1 длиннее фразы 2

phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
Результат:

phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
Результат:
Фразы равной длины

## Программа для Задания №1

```python
# Программа сравнения длин двух фраз
for i in range (3): # Цикл для задания 3х различных условий первой задачи
    if i == 0: # Условие первое
        phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
        phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
    elif i == 1: # Условие второе
        phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
        phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
    else:        # Условие третье
        phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
        phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
    LenPh1 = len (phrase_1) # Определение длины фразы №1
    LenPh2 = len (phrase_2) # Определение длины фразы №2
    if LenPh1 > LenPh2:     # Первая проверка - сравнение длины первой фразы с длиной второй
        print ('Пример №', i+1, ': Фраза 1 длинее, т.к. её длина составляет ', LenPh1, ' против ', LenPh2)
    elif LenPh1 < LenPh2:   # Вторая проверка - сравнение длины второй фразы с длиной первой
        print ('Пример №', i+1, ': Фраза 2 длинее, т.к. её длина составляет ', LenPh2, ' против ', LenPh1)
    else:                   # Остаточное условие при равной длине фраз
        print ('Пример №', i+1, ': Фразы одинаковой длины, т.к. их длина составляет ', LenPh1, ' у 1ой и ', LenPh2, ' у 2ой')

print (' ') # Вывод пустой строки для выделения отступом финальной фразы
print ('Задание №1 выполнено') # Завершение программы
```

## Задание №4
Вам нужно написать программу для подбора упаковок по размерам товара. Размеры хранятся в переменных (в сантиметрах):

width = 10
length = 205
height = 5
Используйте следующие правила:

если каждое из трех измерений менее 15 сантиметров, то выведите на экран “Коробка №1”;
если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите “Коробка №2”;
если длина товара больше 2 метров, то выводите “Упаковка для лыж”;
во всех остальных случаях выводите “Стандартная коробка №3”
## Программа для задания №4

```python
print('Программа для подбора типа упаковки исходя из размеров товара')
print(' ')

# Ввод пользователем габаритов упаковываемого товара
width = int(input('  Введите ширину товара, в сантиметрах: '))
length = int(input('  Введите длину товара, в сантиметрах: '))
height = int(input('  Введите высоту товара, в сантиметрах: '))

# Контрольный вывод введённых габаритов + экранные отступы
print('  Габариты вашего товара (ШДВ):', width, length, height, sep = '/ ', end = ' см')
print(' ')
print(' ')

# --> Основной алгоритм: проверка условий соответствия габаритов товара тазным видам упаковки
if width < 15 and length < 15 and height < 15: # Проверка 1го условия - для коробок 1го типа: любой габарит менее 15 см
    print('Коробка №1')
elif width < 50 and length < 50 and height < 50: # Проверка 2го условия - для коробок 2го типа: любой габарит более 15 см, но все - менее 50 см
    print('Коробка №2')
elif length > 200: # Проверка 3го условия - для лыжной упаковки: длина более 200 см
    print('Упаковка для лыж')
else: # Упаковка для товаров с прочими габаритами
    print('Стандартная коробка №5')
# --> Конец основного алгоритма
    
print(' ') # Вывод пустой строки для выделения отступом финальной фразы
print('Задание №4 выполнено') # Завершение программы
```

## Задание №6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.
Пример работы программы:

Введите тип фигуры:
Круг

Введите радиус круга:
10
Результат:
Площадь круга: 314.16

Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
Результат:
Площадь треугольника: 1.98
## Программа для задания №6

```python
print('Программа для расчёта площади геометрических фигур')
print(' ')  # Вывод пустой строки для выделения отступом

# Выбор типа фигуры для расчёта площади
print('ШАГ №1(2): Выберите тип фигуры, площадь которой необходимо рассчитать:')
print('(1) Круг')
print('(2) Треугольник')
print('(3) Прямоугольник')
print('  ')  # Вывод пустой строки для выделения отступом
choise = int(input('  Введите номер выбранной фигуры: '))

# --> Основной алгоритм: расчёт площади выбранного типа фигур
if choise == 1:                                                # Расчёт площади круга
    print('Считаем площадь круга')
    print(' ')                                                 # Вывод пустой строки для выделения отступом
    radius = int(input('  ШАГ 2(2): Ведите радиус круга, см: ')) # Запрос данных от пользователя
    result = 3,14 * (radius ** 2)                              # Расчёт площади круга
    print(' ')                                                 # Вывод пустой строки для выделения отступом
    print('Площадь круга с радиусом ', radius, ' см составляет ', result, ' см2')
if choise == 2:
    print('Считаем площадь треугольника')
    print(' ')                                                 # Вывод пустой строки для выделения отступом
    A = int(input('  ШАГ 2(2): Ведите длину стороны A, см: ')) # Запрос данных от пользователя 1(3)
    B = int(input('  ШАГ 2(2): Ведите длину стороны B, см: ')) # Запрос данных от пользователя 2(3)
    C = int(input('  ШАГ 2(2): Ведите длину стороны C, см: ')) # Запрос данных от пользователя 3(3)
    P = (A + B + C) / 2                                        # Дополнительный расчёт - полупериметр треугольника
    result = (P * (P - A) * (P - B) * (P - C))**(1/2)          # Расчёт площади треугольника
    print(' ')                                                 # Вывод пустой строки для выделения отступом
    print('Площадь треугольника со сторонами ', A,'/ ', B, '/ ', C, ' см составляет ', result, ' см2')
if choise == 3:                                                # Расчёт площади прямоугольника
    print('Считаем площадь прямоугольника')
    print(' ')                                                 # Вывод пустой строки для выделения отступом
    A = int(input('  ШАГ 2(2): Ведите длину стороны A, см: ')) # Запрос данных от пользователя 1(2)
    B = int(input('  ШАГ 2(2): Ведите длину стороны B, см: ')) # Запрос данных от пользователя 2(2)
    result = A * B                                             # Расчёт площади прямоугольника
    print(' ')                                                 # Вывод пустой строки для выделения отступом
    print('Площадь прямоугольника со сторонами ', A,'/ ', B, ' см составляет ', result, ' см2')    
# --> Конец основного алгоритма
    
print(' ') # Вывод пустой строки для выделения отступом финальной фразы
print('Задание №6 выполнено') # Завершение программы
```

## Задание №5 (необязательное)
Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет “счастливым”. Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Примеры работы программы:

number = 123456
Результат:
Неасчастливый билет

number = 123321
Результат:
Счастливый билет
## Программа для задания №5

```python
print('Программа для анализа номеров проездных билетов')
print(' ')                                                         # Вывод пустой строки для выделения отступом

ticket_number = input('Введите шестизначный номер билета: ')       # Ввод номера билета

print('Проверяемый номер билета: ', ticket_number)
summ1 = ticket_number[0] + ticket_number[1] + ticket_number[2]     # Расчёт суммы первый трёх цифр номера билета
summ2 = ticket_number[-1] + ticket_number[-2] + ticket_number[-3]  # Расчёт суммы вторых трёх цифр номера билета

if summ1 == summ2:                                                 # Проверка номера билета на "счастливость"
    print('Ваш билет счастливый! Съеште его! :)')                  # Информирование о счастливом билете
else:
    print('У вас совершенно обычный билет. После использования - утилизируйте его правильно!')
    
print(' ') # Вывод пустой строки для выделения отступом финальной фразы
print('Задание №6 выполнено') # Завершение программы
```
