# Задание 1
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:

car_id = 'А222BС96’
Результат: Номер А222BС валиден. Регион: 96

car_id = 'АБ22ВВ193’
Результат: Номер не валиден

```python
import re
print('Валидируем введённые номера автомобилей')

car_num_user = input('Введите номер автомобиля [бЦЦЦббЦЦЦ]:')

res = re.findall(r'[авекмнорстухАВЕКМНОРСТУХABCEHKMOPTYX]\d{1,3}[авекмнорстухАВЕКМНОРСТУХABCEHKMOPTYX]{1,2}\d{2,3}', car_num_user)
if res:
    print('Номер валиден', re.findall(r'\w\d{3}\w{2}', car_num_user), '. Регион: ', re.findall(r'(?<=\w\d{3}\w{2})\d{2,3}', car_num_user))
else:
    print('Номер не валиден установленным стандартам. Попробуйте ввести ещё раз')
```

# Задание 3
Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:

some_words = 'Информационные технологии’
Результат: ИТ

some_words = 'Near Field Communication’
Результат: NFC

```python
import re

print('Создаём акронимы (сокращения по первым буквам) для вводимого выражения:')
print('=======================================================================')

user_sentence = input('Введите словосочетание для создания акронима (сокращения): ')

res = re.findall(r'(\b\w)+', user_sentence)
# Шаблон: Повтор (\После начала (края) слова \Любой символ) Любое кол-во повторений

result = ''.join(res)    # Объединяем элементы списка в одно слово (здесь: аббревиатуру)
print('Результат (акроним): ', result)
```

# Задание 6 (необязательное)

Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:
+7-xxx-xxx-xx-xx
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера

```python
import re
print('Проверяем номера телефонов на валидность и приводим к единому формату')
print('=====================================================================')

user_number = input('Введите номер телефона: ')

num_pattern = r'([8]{1}|\+7)(\s|-|\D)(\d{3})(\s|-|\D)(\d{3})(\s|-|\D)(\d{2})(\s|-|\D)(\d{2})'
# Шаблон: (проверка кода РФ)(любой символ-разделитель)(код из 3х чисел)(любой символ-разделитель)
# (Первые три числа)(любой символ-разделитель)(два числа)(любой символ-разделитель)(два числа)
# Проверка на наличие только одной восьмёрки НЕ РАБОТАЕТ!
all_pattern = r'+7 \3-\5-\7-\9'
# Шаблон приведения общего вида: +7 333-333-33-33 Подставляет числа из введённых данных по исходному шаблону

if user_number[0] == '8':                                 # Обход не работающей проверки на проверку множества введёных 8ок
    user_number = re.sub(r'^(\d)+', r'\1', user_number)   # Сброс восьмёрок в начале, приведение к виду 8 ххх... и далее
    
res = re.sub(num_pattern, all_pattern, user_number)       # Приведение введённого номера к единому формату

if res == user_number:                                    # Проверка на валидность
    print('Номер не валиден')                             # Особенность: при несрабатывании шаблонов просто выводит user_number
else:
    print('Номер валиден, приведён к формату: ', res)
```
