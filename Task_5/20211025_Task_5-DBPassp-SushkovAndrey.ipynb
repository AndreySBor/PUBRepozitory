```python
print('Программа для работы с базой данных "аристархов"')
print('================================================')
print()

documents = [                       # Объявление и присвоение базы данных с информацией по документам: тип, номер, владелец
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
]
directories = {                     # Объявление и присвоение значений для базы данных мест хранения документов (полок)
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

command = []                        # Объявдение и обнуление единственной глобальной переменной: для вводимых кодов команд

def main_menu():                    # Функция отображения "Главного меню"
    """
    Функция вывода на экран списка ключей (меню) возможных пользовательских действий и их описания
    Выводится при каждом новом цикле основного алгоритма программы
    """
    print()
    print('Вожможные действия при работе с базой:')
    print('-----------------------------------------')
    print('(q)  Выйти из программы')
    print('(p)  Узнать владельца по номеру документа')
    print('(s)  Узнать по номеру документа его полку')
    print('(l)  Узнать полную информацию о документе')
    print('(m)  Переместить документ на другую полку')
    print('(d)  Удалить документ из данных')
    print('(ad) Добавить новый документ')
    print('(as) Добавить новую полку')
    print('(ds) Удалить имеющуюся полку')
    print()

def unknown_number():              # Функция объявления неудачного поиска документа по его номеру
    """
    Внутренняя служебная функция, обеспечивающая вывод информационного сообщения о:
    1. завершении проверки БД и
    2. отсуствии документа с заданным номером
    """
    print('Результат проверки базы данных: ')
    print('В базе нет документа с таким номером')
    print('----------------------------------------------')

def owner_of_number():            # Функция поиска владельца документа по номеру документа
    """
    Функция, обеспечивающая поиск в БД ФИО владельца документа по номеру документа, вводимого пользователем вручную
    Вызывается ключом 'p'
    """
    print('Ищем владельца документа по номеру документа')# Информационное сообщение для оповещения пользователя о действии
    print('----------------------------------------------')
    number = input('Введите номер документа: ')          # Ввод пользователем номера документа, сохр.в number
    for line in documents:                               # Выделение строки из базы данных (все данные по одному док-ту)
        if number in line['number']:                     # Проверка наличия искомого номера док-та в просмтатриваемой строке БД
            print('Результат проверки базы данных: владелец документа с №', number, ' - ', line['name'])
            print('----------------------------------------------') # Вывод номера с именем владельца при нахождении (+)
            break                                                   # Прерывание цикла, т.к. все номера уникальны
    if number not in line['number']:                                # Проверка условия отсутствия номера в БД
        unknown_number()                                            # Переход к функции Неизвестный номер
    return number, line

def directories_of_number():     # Функция поиска номера полки размещения документа по номеру документа, вводимого пользователем
    """
    Функция, обеспечивающая поиск в БД номера полки документа по номеру документа, вводимого пользователем вручную
    Вызывается ключом 's'
    """
    print('Ищем номер полки по номеру документа')       # Информационное сообщение для оповещения пользователя о действии
    print('----------------------------------------------')
    number = input('Введите номер документа: ')         # Ввод пользователем номера документа, сохр. в number
    for el in directories:                              # Запуск цикла проверки содержимого всех полок хранения
        if number in directories[el]:                   # При (+): найден № док-та на полке №[el]
            print('Результат проверки базы данных: документ с №', number, ' на полке храниения', el) # Инф. сообщение
            print('----------------------------------------------')
            break                                       # Прерывание цикла поиска № док-та в полках
    return number, el                                   # Завершение функции, с выгрузкой № док-та и № полки его находжения
    if number not in directories[el]:                   # Проверка условия отсутствия № док-та
        unknown_number()                                # Вызов функции оповещения об отсутствии документа в БД

def all_in_base():              # Функция вывода полной информации из БД, с доп. пояснениями по маске отображения
    """
    Функция, обеспечивающая поиск в БД номера полки документа по номеру документа, вводимого пользователем вручную
    Вызывается ключом 'l'
    """
    line = {}
    print('Публикация всех имеющихся в базе данных')# Информационное сообщение для оповещения пользователя о действии
    print('----------------------------------------------')
    for line in documents:                          # Выделение строки из базы данных (все данные по одному док-ту)
        number = line['number']                     # Проверка наличия искомого номера док-та в просмтатриваемой строке БД
        for el in directories:                      # Цикл поиска номера полки для активного номера документа
            if number in directories[el]:           # Проверка условия нахождения активного № док-та на el-той полке
                dicrectory = el                     # При (+): сохраниние номера полки, для вывода в результат
                break                               # Прерывание. Проверка на (-) отсутствует, т.к. для всех док-тов д.б.полка
        print('Документ №', number, ', тип:', line['type'], ', владелец:', line['name'], ', полка хранения №', dicrectory)
    print('----------------------------------------------') # Выше: вывод результирующей строки, с доп. пояснениями

def find_directory_number():
    """
    Функция для поиска полки (проверки существования полки в БД)
    Внутренняя служебная функция
    """
    direct_num = input('Введите номер полки для храниения (1, 2, 10 и т.д.):')     # Ввод пользователем номера полки
    num_key = list(directories.keys())              # Выделение всех номеров полок из базы, сохранение в список
    if direct_num in num_key:                       # Проверка условия существования в базе полки с введённым номером
        print('Полка хранения с таким номером существует')  # При (+): подтверждение наличия полки с таким номером
        result = True                               # Присвоение переменной результата проверки
    elif direct_num not in num_key:                 # Проверка условия отсутствия полки с введённым номером
        print('Нет такой полки')                    # При (+): подтверждение отсутствия полки с таким номером
        result = False                              # Присвоение переменной результата проверки
    print('----------------------------------------------')
    return direct_num, num_key, result       # Завершение функции с выгрузкой: № полки, списка со всеми № полок, результатом

def change_direct(command):                         # Функция работы с полками хранения
    """
    Функция для добавления и удаления полок хранения.
    Вызов ключом 'as' или 'ds'
    """
    direct_num, num_key, fRes = find_directory_number() # Запуск поиска № полки, с выгрузкой: № полки, списком полок, рез-том
    
    if command == 'as' and direct_num not in num_key: # Проверка условий для добавления новой полки хранения: команда и нет в БД
        print('Создаю новую полку хранения с номером ', direct_num) # Сообщение о проводимой операции
        directories[direct_num] = []                  # Создание в БД (словаре) новой полки (ключ с введённым номером)
        print('Новый перечень полок хранения и их содержимое:') # Контрольный вывод 
        print(directories)                            # Контрольный вывод БД со всеми полками и их содержимым
    elif command == 'ds' and direct_num in num_key:   # Проверка условий для удаления полки: команда и наличие в БД (словаре)
        print('Содержимое удаляемой полки:', directories[direct_num]) # Контрольный вывод содержимого удаляемой полки
        print('Переношу содержимое в полку хранения №', num_key[0])   # Предупреждение о переносе док-тов в полку №...
        print('Выбранная операция не удаляет документы. При необходимости удалить все документы в полке хранения, используйте другую команду')
        directories[num_key[0]] = directories[num_key[0]] + directories[direct_num] # Добавление содежимого удаляемой полки к 1ой полке словаря
        print('Удаляю полку хранения с номером ', direct_num)     # Контрольный вывод сообщения об операции
        del(directories[direct_num])                              # Удаление полки (удаление ключа из словаря)
        print('Новый перечень полок хранения и их содержимое:')   # Контрольный вывод
        print(directories)                                        # Содержимое обновлённой БД (словаря полок)
    else:                                                         # Сообщение о невозможности проведения запрошенных действий
        print('Указанное действие применить к текущей версии БД невозможно')

def add_doc():                                  # Функция добавления нового документа в БД
    """
    Функция для добавления нового документа.
    Вызывается ключом 'ad'
    """
    mem_dict = []                               # Объявление и обнуление списка для сохранения данных полок
    line = {}                                   # Объявление и обнуление словаря для получения данных из др. функций

    number, line = owner_of_number()            # Проверка наличия номера док-та в БД, получение № док-та и сроки с номерами

    if number not in line['number']:                         # Проверка условия отсутствия номера в БД
        type_doc = input('Введите тип документа (passport/invoice/insurance/multipassport/etc.):') # Ввод типа док-та
        owner_doc = input('Введите ФИО или имя владельца документа:')                              # Ввод имени владельца
        
        direct_num, num_key, fRes = find_directory_number()  # Алгоритм запроса № полки хранения, её создания при отсутствии
        
        if fRes == False:                                    # При отсутствии введённого № полки в базе
            change_direct('as')                              # Запрос на функцию создания новой полки, с командой на создание

        mem_dict = directories[direct_num]                   # Сохранение содержимого полки в промежуточный список
        mem_dict.append(number)                              # Добавление нового № документа к содержимому полки
        directories[direct_num] = mem_dict                   # Возврат содержимого полки в основную БД (словарь)
        
        documents.append({'type':type_doc, 'number':number, 'name':owner_doc}) # Добавление данных нового док-та в осн. БД
        print()
        print('Добавлен новый документ с параметрами:')      # Вывод контрольного сообщения о завершении операции добавления
        print('Документ №', number, ', тип:',type_doc, ', владелец:', owner_doc, ', полка хранения №', direct_num)
        print(directories)                                   # Вывод всех полок с содержимым (номера док-тов)
        all_in_base()                                        # Вывод всего содержимого БД в развёрнутой форме

def del_doc():
    """
    Функция для удаления документа из БД.
    Вызывается ключом 'd'
    """
    print()
    print('Выполняем удаление документа из базы')            # Информационное сообщение о выполняемой операции
    number, line = owner_of_number()                # Проверка наличия номера док-та в БД, получение № док-та и его др.данных
    if number in line['number']:                             # Проверка условия наличия номера в БД
        print('Удалаяем документ с №', number, '. Параметры документа:') # Вывод контрольного сообщения с данными документа
        print(line)                                          # Вывод контрольного сообщения с данными документа
        index_doc = documents.index(line)                    # Определение индекса данных документа в БД (список со словарями)
        del(documents[index_doc])                            # Удаление словаря с данными док-та, по индексу в списке
        print('Удаление завершено')                          # Вывод инфо-сообщения о заврешении операции
        all_in_base()                                        # Контрольный вывод содержимого БД после удаления док-та
    
    elif number not in line['number']:                       # Контрольное условие: отсутствие запрошенного номера в БД
        unknown_number()                                     # Вывод информации об отсутствии док-та с запрошенным номером

def trans_doc():
    """
    Функция для перемещения документа из одной полки хранения в другую.
    Вызывается ключом 'm'
    """
    print()
    print('Выполняем перемещение документа из полки в другую полку хранения') # Информационное сообщение о выполняемой операции
    number, old_direct = directories_of_number()             # Поиск № док-та в БД, выгрузка в переменные № док-та и № исходной полки
    print('Сейчас документ в полке', old_direct)             # Контрольный вывод номера текущей полки хранения документа
    new_direct = input('Введите номер полки, куда перенести документ:') # Ввод пользователем номера новой полки
    
    if new_direct not in directories.keys():                 # Проверка введённоого номера полки на существование в БД
        change_direct('as')                                  # При (-): запуск функции создания новой полки хранения через код
    
    line = directories[old_direct]                           # Присвоение значений № док-тов, хранящихся в исходной полке
    
    for el in line:                                          # Цикл поиска № док-та в исходной полке
        if el == number:                                     # При (+): нахождение № док-та
            line.remove(el)                                  # Удаление № док-та в исходной полке (значение в списке)
            break                                            # Прерывание цикла поиска № док-та в исходной полке
    directories[old_direct] = line                           # Перезапись в словарь (БД) изменённого перечня №№ док-тов
    print('After delete', directories)                       # Контрольный вывод содержимого полок хранения
    
    line = directories[new_direct]                           # Перезапись содержимого новой полки в рабочую переменную (список)
    line.append(number)                                      # Добавление № док-та к содержимому новой полки хранения
    directories[new_direct] = list(line)                     # Запись дополненного содержимого в новую полку хранения
    print('After getting', directories)                      # Контрольный выврд содержимого полок хранения
    print('Операция переноса завершена')                     # Инф. сообщение о завершении операции переноса документа
    
    # ОСНОВОЙ АЛГОРИТМ: замкнутый цикл while, ожидающий ввода пользователем ключа для действия. Выход - только по 'q'
while command != 'q':                                        # Проверка условия завершения цикла (программы): ввод 'q'
    main_menu()                                              # Вызов функции отображения Главного пользовательского меню
    command = input('Введите код необходимого действия: ')   # Запрос пользователю на ввод ключа команды для исполнения
    if command == 'q':                                       # Доп. проверка условия для прерывания цикла / программы
        break                                                # Прерывание исполнения программы
    elif command == 'p':                                     # Проверка условия вызова команды для определения владельца по №
        owner_of_number()                                    # Вызов функции поиска владельца по номеру документа
    elif command == 's':                                     # Проверка усл.вызова команды для определения полки по № док-та
        directories_of_number()                              # Вызов функции определения № полки по номеру док-та
    elif command == 'l':                                     # Проверка усл. вызова функции вывода полной информации из БД
        all_in_base()                                        # Вызов функции вывода полной информации из БД
    elif command == 'as':                                    # Проверка усл. вызова функции добавления новой полки
        change_direct(command)                               # Вызов функции работы с полками хранения (добавление/удаление)
    elif command == 'ds':                                    # Проверка усл. вызова функции удаления имеющейся полки
        change_direct(command)                               # Вызов функции работы с полками хранения (добавление/удаление)
    elif command == 'ad':                                    # Проверка усл. вызова функции добавления нового документа
        add_doc()                                            # Вызов функции для добавления нового документа
    elif command == 'd':                                     # Проверка усл. вызова функции удаления имеющегося документа
        del_doc()                                            # Вызов функции удаления документа
    elif command == 'm':                                     # Проверка усл. вызова функции перемещения документа
        trans_doc()                                          # Вызов функции перемещения документа между полками
    else:                                                    # Poka-yoka: проверка на некорректный ввод кода команды
        print('Введён некорректный код. Пожалуйста, повторите свой выбор')
    # КОНЕЦ ОСНОВНОГО АЛГОРИТМА
        
print('================================================')
print('Программа завершена')
```
