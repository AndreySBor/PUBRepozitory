## Задание 1
Дана переменная, в которой хранится словарь, содержащий гео-метки для каждого пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран множество уникальных гео-меток всех пользователей.

Пример работы программы:

ids = {‘user1’: [213, 213, 213, 15, 213],
‘user2’: [54, 54, 119, 119, 119],
‘user3’: [213, 98, 98, 35]}
Результат: {98, 35, 15, 213, 54, 119}

## Программа к заданию №1

```python
print('Вывод уникальных метрик из словаря')
print('==================================')

result = list()                              # Объявление и обнуление списка для сохранения результатов

ids = {                                      # Объявление справочника, с присвоением значений
    'user1': [213, 213, 213, 15, 213],
    'user2': [54, 54, 119, 119, 119],
    'user3': [213, 98, 98, 35]
    }

for user in ids.keys():                      # Основной алгоритм: цикл по количеству уникальных метрик в словаре
    log_list = list(ids[user])               # Присвоение промежуточному списку значений для текущего пользователя
    result = result + list(log_list)         # Прибавление метрик текущего пользователя к общему перечню метрик

print('----------- Результат ------------')
print(set(result))                           # Вывод результата + убираем дубликаты: перевод типа "список" в тип "справочник"
print('==================================')
print('Программа заврешена')
```

## Задание 2
Дана переменная, в которой хранится список поисковых запросов пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

Пример работы программы:

queries = [
‘смотреть сериалы онлайн’,
‘новости спорта’,
‘афиша кино’,
‘курс доллара’,
‘сериалы этим летом’,
‘курс по питону’,
‘сериалы про спорт’,
]
Результат:

Поисковых запросов, содержащих 2 слов(а): 42.86%
Поисковых запросов, содержащих 3 слов(а): 57.14%

## Программа к заданию №2

```python
print('Программа подсчёта количества слов в поисковых запросах')
print('=======================================================')

queries = [                                                # Ввод N-поисковых запросов из произвольного кол-ва слов
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт'
]

word = 0                    # Присвоение и обнуление переменной для счёта кол-ва слов в запросе
res_1 = []                  # Объявление и обнуление списка для сохранения свех длин запросов
queries_type_base = {}      # Объявление и обнуление словаря для сохранения уникальных количеств слов и результатов их подсчёта

for word in queries:                                     # Основной алгоритм: запуск цикла подсчёта кол-ва слов в запросах
    length = int(len(word.split(' ')))                   # Присвоение переменной длины запроса, в словах
    res_1.append(length)                                 # Сохранение результата измерения длины запроса в списке

    if length in queries_type_base:                      # Проверка наличия найденной длины запроса в уже найденных ранее
        quantity = int(queries_type_base[length])        # Условие TRUE: вынимаем ранее сохранённое в справочнике кол-во запросов для найденной длины
        quantity += 1                                    # Прибавляем к сохранённому кол-ву ещё один найденный запрос
        queries_type_base[length] = quantity             # Перезаписывем в справочник новое кол-во запросов с данной длиной
    else:                                                # Условие FALSE:
        queries_type_base[length] = 1                    # Для ранее ещё не найденной длины заводим ячейку в справочнике со значением кол-ва =1
        
print('----------- Результат ------------')              # Раздел для расчёта и вывода результатов
print('Идентифицированы запросы с количеством слов: ',(list(queries_type_base.keys()))) # Вывод перечня найденных длин запросов
print()                                                  # Пропуск строки
part_1 = list(queries_type_base.keys())                  # Отделение списка найденных вариантов длин запросов, шт. слов
part_2 = list(queries_type_base.values())                # Отделение списка кол-ва найденных запросов для каждой из длин слов, шт.

# print('Keys: ', part_1)
# print('Values: ', part_2)

summary, i, ii = [0, 0, 0]                               # Объявление и обнуление дополнительных переменных для расчётов

for i in part_2:                                         # Цикл для расчёта общей суммы (кол-ва) запросов - для вычисления долей
    summary = summary + int(i)                           # Расчёт общей суммы (кол-ва) запросов - для вычисления долей

while ii < len(part_1):                                  # Цикл для расчёт и вывода результатов - произвольное кол-во
    print('Поисковых запросов, содержащих', part_1[ii], 'слов(а):', ((part_2[ii] / summary) * 100), '%') # Расчёт и вывод долей для каждой из длин запросов
    ii += 1                                              # Служебная переменная: переход цикла к следующему результату

print('==================================')
print('Программа заврешена')
```

## Задание 3
Дана переменная, в которой хранится информация о затратах и доходе рекламных кампаний по различным источникам. Необходимо дополнить исходную структуру показателем ROI, который рассчитаем по формуле: (revenue / cost - 1) * 100

Пример работы программы:

results = {
‘vk’: {‘revenue’: 103, ‘cost’: 98},
‘yandex’: {‘revenue’: 179, ‘cost’: 153},
‘facebook’: {‘revenue’: 103, ‘cost’: 110},
‘adwords’: {‘revenue’: 35, ‘cost’: 34},
‘twitter’: {‘revenue’: 11, ‘cost’: 24},
}
Результат:

{‘adwords’: {‘ROI’: 2.94, ‘cost’: 34, ‘revenue’: 35},
‘facebook’: {‘ROI’: -6.36, ‘cost’: 110, ‘revenue’: 103},
‘twitter’: {‘ROI’: -54.17, ‘cost’: 24, ‘revenue’: 11},
‘vk’: {‘ROI’: 5.1, ‘cost’: 98, ‘revenue’: 103},
‘yandex’: {‘ROI’: 16.99, ‘cost’: 153, ‘revenue’: 179}}
## Программа к заданию №3

```python
print('Программа подсчёта количества слов в поисковых запросах')
print('=======================================================')
# Программа для извлечения данных из словаря, их обработки и добавления результатов к исходному словарю

results = {                                              # Присвоение основному словарю исходных данных
    'vk': {'revenue': 103, 'cost': 98}, 
    'yandex': {'revenue': 179, 'cost': 153}, 
    'facebook': {'revenue': 103, 'cost': 110}, 
    'adwords': {'revenue': 35, 'cost': 34}, 
    'twitter': {'revenue': 11, 'cost': 24}, 
}

for company in results:                                  # Основной цикл: выборка словарей отдельной компании из осн.словаря
    cost = results[company]['cost']                      # Присвоение переменной исходных данных затрат: для расчёта и возврата в исходный словарь
    revenue = results[company]['revenue']                # Присвоение переменной исходных данных дохода: для расчёта и возврата в исходный словарь
    ROI = (revenue / cost - 1) * 100                     # Расчёт показателя ROI
    print(company, 'ROI:', ROI)                          # Контрольныый вывод на экран расчёта ROI
    results[company] = ['ROI:', ROI, 'revenue:', revenue, 'cost:', cost] # Возврат всех данных компании в исходный словарь

print('----------- Результат ------------')              # Раздел для расчёта и вывода результатов
print(results)                                           # Вывод результата на экран
print('==================================')
print('Программа заврешена')
```

## Задание 4
Дана переменная, в которой хранится статистика рекламных каналов по объемам продаж (пример структуры данных приведен ниже). Напишите программу, которая возвращает название канала с максимальным объемом продаж.

Пример работы программы:

stats = {‘facebook’: 55, ‘yandex’: 115, ‘vk’: 120, ‘google’: 99, ‘email’: 42, ‘ok’: 98}

Результат: Максимальный объем продаж на рекламном канале: vk
## Программа для задания №4

```python
print('Программа оценки данных максимальных объёмов продаж')
print('===================================================')
# Программа оценки максимальных значений словаря

stats = {                      # Присвоение основному словарю исходных данных
    'facebook': 55,
    'yandex': 115,
    'vk': 120,
    'google': 99,
    'email': 42,
    'ok': 98
}

company = []                   # Объявление и обнуление перемонной для наименования компаний
maximum_items = []             # Объявление и обнуление переменной для сохранения результата: компании с max показетелем
max_val = 0                    # Объявление и обнуление перерменой для сохранения сравниваемых показателей компаний

for company in stats:          # Основной алгоритм: цикл просмотра исх.данных - словаря с компаниями их показателями - по имени
    if max_val < stats[company]:         # Проверка условия: max найденный показатель к текущему проверяемому (выход при (-))
        max_val = stats[company]         # (+): Присвоение max найденного текущего значения показателя
        maximum_items = str(company)     # (+): Присвоение результирующей переменной наименования текущей проверяемой компании

print('----------- Результат ------------')                                 # Раздел для расчёта и вывода результатов
print('Максимальный объем продаж на рекламном канале: ', maximum_items)     # Вывод результата на экран
print('==================================')
print('Программа заврешена')
```

## Задание 5 (необязательно)
Дан список произвольной длины. Необходимо написать код, который на основе исходного списка составит словарь такого уровня вложенности, какова длина исхондого списка.

Примеры работы программы:

my_list = [‘2018-01-01’, ‘yandex’, ‘cpc’, 100]
Результат: {‘2018-01-01’: {‘yandex’: {‘cpc’: 100}}}

my_list = [‘a’, ‘b’, ‘c’, ‘d’, ‘e’, ‘f’]
Результат: {‘a’: {‘b’: {‘c’: {‘d’: {‘e’: ‘f’}}}}}
## Программа для задания №5

```python
print('Программа для создания вложенных словарей, исходя из исходных данных')
print('====================================================================')
# Программа создаёт словарь с уровнем вложенных словарей -1 от кол-ва исходных данных

variant = int(input('Введите номер проверяемой базы (1 или 2):')) # Ввод пользователем варианта наполнения базы
if variant == 1:                                                  # Проверка варианта №1
    my_list = ['2018-01-01', 'yandex', 'cpc', 100]                # Объявление и присвоение исх.данных рабочему словарю
elif variant == 2:                                                # Проверка варианта №2
    my_list = ['a', 'b', 'c', 'd', 'e', 'f']                      # Объявление и присвоение исх.данных рабочему словарю
else:                                                             # Poka-yoka: защита от ввода неправильных значений
    print('Необходимо повторить выбор, т.к. вариант не 1 или 2')

print('Выбранная последовательность данных:', my_list)            # Контрольный вывод: анализируемая последовательность

x = {}                                             # Объявление и обнуление промежуточного словаря
res_5 = {}                                         # Объявление и обнуление результирующего словаря
length = len(my_list)                              # Запись в переменную кол-ва составляющих анализируемой последовательности
i = int(length - 1)                                # Объявление переменной для цикла: со сдвигом на (-1)

while i >= 0:                                              # Основной цикл: обратный, от (кол-ва элементов -1) до нуля
    if i == int(length - 1):                               # Проверка условия: идентификация первого круга цикла
        x[my_list[length - 2]] = my_list[length - 1]       # Присвоение первого значения словаря - ключ и зн-ие, с конца последовательности (согласно условий задания)
    elif i != int(length - 1) and i != int(length - 2):    # Проверка условия на несовпадение с первым (т.к есть 1ый If) и вторым циклом (т.к.в 1ом цикле записаны сразу 2 значения из основной последовательности)
        res_5 = {}                                         # ОБЯЗАТЕЛЬНО! Обнуление словаря результата
        res_5.setdefault(my_list[i], x)                    # Присвоение очередного ключа из осн.последовательности + значения: записанного ранее словаря
        x = res_5                                          # Перезапись полученного результата в промежуточный словарь (для добавления в следующем цикле)
    i -= 1                                                 # Переход к следующему шагу цикла (-1)

print('----------- Результат ------------')                    # Раздел для вывода результатов
print(res_5)                                                   # Вывод результата на экран
print('==================================')
print('Программа заврешена')
```
