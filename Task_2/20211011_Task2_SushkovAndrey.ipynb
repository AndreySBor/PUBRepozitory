## Задание №1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

среднюю букву, если число букв в слове нечетное;
две средних буквы, если число букв четное.
Примеры работы программы:

word = ‘test’
Результат:
es

word = ‘testing’
Результат:
t
## Программа для задания №1

```python
print('Программа для поиска средних букв в слове')
print('--------------------------------')

word_input = str(input('  Введите любое слово латинскими буквами: '))
len_word = len(word_input)
if len(word_input) % 2 == 0:
    print('  В слове чётное число букв')
    fst_letter = int(len(word_input) / 2)
    print('  Средние буквы: ', word_input[fst_letter-1:fst_letter+1])
else:
    print('  В слове НЕчётное число букв')
    fst_letter = int(len(word_input) / 2)
    print('  Средняя буква: ', word_input[fst_letter:fst_letter+1:2])

print('_________________________________')    
print('Программа завершена')
```

## Задание №2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы:

Введите число:
1

Введите число:
4

Введите число:
6

Введите число:
0
Результат:
11

Введите число:
0
Результат:
0
## Программа для задания №2

```python
print('Программа суммирования по условию')
print('--------------------------------')

i = 2
result = 0
ent_int = int(input('Ввод числа №1. Введите любую цифру в диапазоне 0-9: '))
while ent_int != 0:
    result = result + ent_int
    print('Ввод числа №', i)
    ent_int = int(input('Введите любую цифру в диапазоне 0-9: '))
    i +=1

print('_________________________________') 
print('Сумма введённых чисел равна ', result)
print('_________________________________')    
print('Программа завершена')
```

## Задание №3
Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек.
Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки! Но мы не будем никого знакомить, если кто-то может остаться без пары:

Примеры работы программы:

boys = [‘Peter’, ‘Alex’, ‘John’, ‘Arthur’, ‘Richard’]

Результат:

## Программа для задания №3

```python
print('Программа для подбора "идеальной пары"')
print('--------------------------------------')

variant = int(input('Выберите ситуацию: 1 или 2 ->')) # Пользователь выбирвает возможный вариант по количеству человек
if variant == 1:
    boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
    girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']
elif variant == 2:
    boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Michael']
    girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']    
else:                                                  # Проверка на ошибку ввода некорректных значений
    print('Ввели чего-то не то, нужно попробовать ещё раз')

if len(boys) != len(girls):                            # Сравнение количества boys & girls  в списках
    print('Внимание, кто-то может остаться без пары!') # Остановка подбора пар из-за несоответствия количества участников разных полов
else:
    pairs = zip(sorted(boys), sorted(girls))           # Основной алгоритм: сортировка списков по алфавиту и их совмещение
    print('--------------------------------------')
    print('Идеальные пары:', list(pairs))              # Трансформация кортежа в список для вывода результата на экран

print('--------------------------------------')
print('Программа завершена')
```

## Задание №4
У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере). Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

Пример работы программы:

countries_temperature = [
[‘Thailand’, [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
[‘Germany’, [57.2, 55.4, 59, 59, 53.6]],
[‘Russia’, [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
[‘Poland’, [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
Результат:

Средняя температура в странах:
Thailand - 23.9 С
Germany - 13.8 С
Russia - 3.7 С
Poland - 12.0 С
## Программа для задания №4

```python
print('Программа для оперделения среднегодовой температуры')
print('===================================================')

countries_temperature = [['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]], ['Germany', [57.2, 55.4, 59, 59, 53.6]], ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]], ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]]

num_country = 0                                         # Контрольное обнуление: порядкового номера для первого значения в основном списке
for country in countries_temperature:                   # Запуск цикла для проходов по списку с исходными данными, по автоопределяемому количеству стран в списке с исходными данными
    Country = countries_temperature[num_country]        # Шаг 1: Выделение данных по конкретной стране из общего списка с исходными данными
    country_temp = Country[1]                           # Шаг 2: Выделение из данных по стране списка с годовыми температурами
    mid_temp_Country, i = [0, 0]                        # Контрольное обнуление переменных: для цикла пересчёта стат.данных по температурам
    for Ctemp in country_temp:                          # Запуск цикла: для произвольного количества имеющихся в исходном списке температур
        Ctemp = (country_temp[i] - 32) * (5 / 9)        # Шаг 3: Перевод шкалы Фаренгейта в шкалу Цельсия
        mid_temp_Country = mid_temp_Country + Ctemp     # Шаг 4: Расчёт суммы годовых температур в шкале Цельсия
        i+=1                                            # Счёт переменной для знаменателя расчёта средней годовой температуры
    print('Средняя годовая температура воздуха в ',Country[0], 'равна', mid_temp_Country/i, 'С') # Шаг 5: Вывод результата с расчётом средней температуры, с автоподстановкой названия страны из исходного списка
    num_country +=1                                     # Переход к следующей стране в основном списке
    print('--------------------------------------')

print('===================================================')
print('Программа завершена')
```

## Программа для задания №5

```python
print('Программа для подсчёта просмотренных страниц по логам конкретных пользователей')
print('==============================================================================')

variant = int(input('Выберите ситуацию: 1 или 2 ->'))  # Пользователь выбирвает возможный вариант
if variant == 1:                                       # Присвоение списку исходных данных по варианту №1
    stream = ['2018-01-01,user1,3', '2018-01-07,user1,4', '2018-03-29,user1,1', '2018-04-04,user1,13', '2018-01-05,user2,7', '2018-06-14,user3,4', '2018-07-02,user3,10', '2018-03-21,user4,19', '2018-03-22,user4,4', '2018-04-22,user4,8', '2018-05-03,user4,9', '2018-05-11,user4,11']
elif variant == 2:                                     # Присвоение списку исходных данных по варианту №2
    stream = ['2018-01-01,user100,150', '2018-01-07,user99,205', '2018-03-29,user1001,81']    
else:                                                  # Проверка на ошибку ввода некорректных значений при выборе варианта
    print('Ввели чего-то не то, нужно попробовать ещё раз')

log_number, i, user_veiw, users_sum = [0, 0, 0, 0]     # Объявление переменных типа Integer, с контрольным обнулением
users_su, users = ['','']                              # Объявление переменных типа String, с контрольным обнулением

for log_number in stream:                              # Запуск цикла анализа исходных данных, по количеству логов в исходном списке
    log = stream[i]                                    # Выделение единичного лога из исходного списка, по порядковому номеру
    log_value = log.split(',')                         # Разделение единичного лога на составляющие, по разделителю ","
    # print(log_value)                                   # Контрольный вывод лога на экран
    user_veiw = user_veiw + int(log_value[2])          # Счёт общей суммы просмотров, с выделением кол-ва просмотров из единичного лога
    
    users = log_value[1]                               # Присвоение имени пользователя из текущего анализируемого лога
    if users not in users_su:                          # Проверка присутствия уникального имени пользователя в рабочнм списке уникальных имён цикла
        users_su = users_su + users                    # Добавление имени пользователя к списку уникальных имён цикла
    i+=1                                               # Сдвижка переменной для перехода к анализу следующего лога (след. цикл)

users_sum = users_su.count('user')                     # Счёт уникальных пользователей
print('---------------------------------------------------')
print('      Контрольные суммы:')
print('Всего уникальных пользователей: ', users_sum)   # Вывод контрольной суммы: уникальные пользователи
print('Всего просмотров: ', user_veiw)                 # Вывод контрольной суммы: сумма просмотров
print('---------------------------------------------------')

# Вывод основного результата работы программы: с расчётом среднего кол-ва просмотров на одного уникального пользователя
print('Среднее количество просмотров на уникального пользователя: ', (user_veiw / users_sum))

print('===================================================')
print('Программа завершена')
```

## Задание №6 (необязательное)

Примеры работы программы:

Введите числа:
4 8 0 3 4 2 0 3
Результат:
0 3 4

Введите числа:
1 1 2 2 3 3

1 2 3

Введите числа:
1 1 1 1 1 2 2 2

1 2
## Программа для задачи №6

```python
print('Программа для поиска дубликатов введённых пользователем данных (чисел)')
print('======================================================================')

user_set = input('Введите набор чисел через "пробел" (ex.: 2 8 6 3): ')  # Ввод пользователем последовательности для анализа

i, x = [0, 0]                                                            # Объявление служебных переменных Integer, с обнулением
number, number_set, result = ['','','']                                  # Объявление служебных переменных String, с обнулением

number_set = user_set.split(' ')                                         # Разделение введённой последовательности на отдельные числа
for x in number_set:                                                     # Запуск цикла для анализа состава последовательности
    number = number_set[i]                                               # Выделение единичной части последовательности
    if number_set.count(number) > 1:                                     # Проверка условия повторяемости части в последовательности
        if number not in result:                                         # Проверка условия учёта найденного дубликата в результирующем списке
            result = result + ' ' + number                               # Добавление идентифицированного дубликата к конечному результирующему списку
    i+=1                                                                 # Переход к слеующему элементу последовательности (цикла)

print('Дубликаты: ', result)                                             # Вывод конечного результата
print('======================================================================')
print('Программа завершена')
```
